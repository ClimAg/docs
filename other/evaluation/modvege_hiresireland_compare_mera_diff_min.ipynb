```python
from datetime import datetime, timezone
import climag.plot_stats as cstats
import importlib
```

```python
season_list = ["DJF", "MAM", "JJA", "SON"]
```

```python
data = cstats.hist_obs_diff(stat="min", dataset="HiResIreland")
```
